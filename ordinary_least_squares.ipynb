```python
%matplotlib widget
import numpy as np
import pandas as pd
import scipy 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
demographics_df = pd.read_csv('demographics.csv')
features = demographics_df.columns[2:-3]
X = demographics_df[features]
y = demographics_df["margin"]
```

```python
X_standardized = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, random_state=0)
model: LinearRegression = LinearRegression().fit(X_train, y_train)

y_test_pred = model.predict(X_test) 

r_sqr_train = model.score(X_train, y_train)
r_sqr_test = model.score(X_test, y_test)

print("R-squared (Train Set): {}".format(r_sqr_train))
print("R-squared (Test Set): {}".format(r_sqr_test))
```

```python
linear_regression_df = demographics_df.assign(pred_margin=model.predict(X_standardized))
linear_regression_df
```
